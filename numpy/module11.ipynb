### Arithmetic Operation and mathematical function

```python
#trigonometry
a= np.random.randint(0,10, size = (1,3))

# sin, cos, tan, degree takes values in radian
sin_val = np.sign(a)
cos_val = np.cos(a)
tan_val = np.tan(a)
deg_con = np.rad2deg(a)

print(f"{a} \n")
print(f"{sin_val}\n")
print(f"{cos_val} \n")
print(f"{tan_val}\n")
print(f"{deg_con}\n")



```

# Searching

```python
import numpy as np 
x = np.array([10,8,6,25])
print(x)
# np.where(condition, x, y ) if ture then x or if not then replace with y
print(np.where(x==6))
print(np.where(x>6, x, 0))


print(np.where(mat>6, mat, 0))

max_val = np.argmax(x)
min_val = np.argmin(x)
print(max_val)
print(min_val)
```

## Statistical methods

```python
#data load 

data = np.genfromtxt("./data/student_scores.csv", delimiter=",", skip_header=1)
print(data)

```

```python
#statistics

math_marks = data[:,:3]

print(math_marks.T)

max_math_marks = np.max(math_marks)
min_math_marks = np.min(math_marks)
print(max_math_marks)
print(min_math_marks)
print(np.mean(math_marks))
print(np.median(math_marks))
print(np.std(math_marks))
```

```python
#corrcoef  
study_time = np.array([2,3,4,5,6])
marks = np.array([70,75,80,85,90])
correlation = np.corrcoef(study_time, marks)
print(correlation)
```

## Linear Algebra

```python
A = np.array([[1,2,3],[4,5,6]])
B = np.array([[7,8],[9,10],[11,12]])

C = np.dot(A, B)
print(C)
print(np.trace(B))
```

```python
sq_mat = np.array([[1,2,9],[3,4,2],[5,6,8]])
print(np.shape(sq_mat))

det_of_sq_mat = np.linalg.det(sq_mat)
print(det_of_sq_mat)

renk = np.linalg.matrix_rank(sq_mat)
print(renk)
```
