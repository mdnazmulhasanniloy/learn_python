```python
number = input("Enter numbers separated by space: ") 
numbers_list = [int(x) for x in number.split()]
result = []
for num in numbers_list:
    if num % 3 == 0 and num % 5 != 0:
        result.append(num)

print(result)
```

```python
def compute(a, b):
    try: 
        squire = lambda x: x**2
        result = (squire(a) + squire(b)) / (a - b);
        return result
    except ZeroDivisionError: 
        return "Error: Division not possible"


a, b = map(int, input("Enter 2 numbers separated by space: ").split())

print(compute(a, b))
```

```python
import string 
from collections import Counter  
 

def Content(filename):
    with open (filename, 'r') as file:
        text = file.read().lower()
        text = text.translate(str.maketrans("", "", string.punctuation))
        text = text.replace("\n", " ")
        words = text.split()
        mostCommonWords = Counter(words)
        sortad = sorted(mostCommonWords.items(), key=lambda x: x[1], reverse=True)
        return sortad[0][0]
 
 
print(Content("data.txt"))
```

```python
from collections import Counter

def convert_sentence_into_list(sec):
    sec = sec.split()
    result = [] 

    
    for i in range(len(sec)):
        letters = len(sec[i])
        result.append(letters)

    return result


sentence = input("Enter a sentence: ")
print(convert_sentence_into_list(sentence))
```

```python
def checkMirror (number):
    if str(number) == str(number)[::-1]:
        return "Mirror Number"
    else:
        return "Not Mirror Number" 
    
text = input("Enter a number: ")   

print(checkMirror(text))
```
